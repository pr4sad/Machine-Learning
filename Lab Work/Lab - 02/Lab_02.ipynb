```python
def compute_model_output(x, w, b):
      m = x.shape[0]
      f_wb = np.zeros(m)
      for i in range(m):
        f_wb[i] = w * x[i] + b

      return f_wb
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function to optimize: f(x) = x^2
def f(x):
    return x**2

# Define the derivative of the function: f'(x) = 2x
def df(x):
    return 2*x

# Cost function to evaluate the model's performance
def cost_function(x, y, w, b):
    n = x.shape[0]  # Number of data points
    cost = 0  # Initialize cost

    # Compute the cost based on the model's predictions
    for i in range(n):
        j_wb = w * x[i] + b  # Linear prediction (w*x + b)
        diff = j_wb - y[i]  # Difference between prediction and actual value
        cost += (diff)**2  # Accumulate the squared differences

    total_cost = 1 / (2 * n) * cost  # Average cost
    return total_cost

# Gradient descent with momentum implementation
def gradient_descent_momentum(x_train, y_train, learning_rate=0.1, momentum=0.9, num_iterations=100, initial_x=5, velocity=0):
    x = initial_x  # Starting point for optimization
    x_values = [initial_x]  # List to store x values
    f_values = [f(initial_x)]  # List to store function values

    # Perform gradient descent iterations
    for i in range(num_iterations):
        gradient = df(x)  # Calculate the gradient at current x
        velocity = momentum * velocity - learning_rate * gradient  # Update velocity
        x = x + velocity  # Update x using velocity

        # Store values for plotting
        x_values.append(x)
        f_values.append(f(x))

        # Print the progress for each iteration
        print(f"Iteration {i+1}: x = {x}, f(x) = {f(x)}")

    print(f"Final value: x = {x}, f(x) = {f(x)}")  # Print final result

    return x_values, f_values  # Return values for plotting

# Initialize parameters for gradient descent
learning_rate = 0.1
momentum = 0.9
num_iterations = 100
initial_x = 5
velocity = 0

# Example training data (x_train and y_train)
x_train = np.array([1.0, 2.0, 3.0, 5.0, 8.0])
y_train = np.array([300.0, 500.0, 800.0, 900.0, 500.0])

# Execute gradient descent
x_values, f_values = gradient_descent_momentum(x_train, y_train, learning_rate, momentum, num_iterations, initial_x, velocity)

# Plot the results of the optimization process
plt.plot(x_values, f_values, marker='o')
plt.xlabel('x')  # X-axis label
plt.ylabel('f(x)')  # Y-axis label
plt.title('Gradient Descent Optimization with Momentum')  # Plot title
plt.show()  # Display the plot

# Compute and print the final cost using the cost function
w = 0.1  # Example value for weight
b = 0.1  # Example value for bias
cost = cost_function(x_train, y_train, w, b)  # Calculate cost
print(f"Final cost: {cost}")  # Print final cost
```
